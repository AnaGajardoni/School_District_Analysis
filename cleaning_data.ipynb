```python
# Substitute the missing math scores with the mean of math scores for the ones that actually have a value

missing_grade_df['math_score'] = missing_grade_df['math_score'].fillna(missing_grade_df.describe().loc['mean'].math_score)
missing_grade_df
```

```python
# Substitute the missing reading scores with the mean of reading scores for the ones that actually have a value

missing_grade_df['reading_score'] = missing_grade_df['reading_score'].fillna(missing_grade_df.describe().loc['mean'].reading_score)
missing_grade_df
```
